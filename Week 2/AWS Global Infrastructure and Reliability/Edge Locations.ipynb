### **Edge Locations**

Một địa điểm biên **là một trang web mà Amazon CloudFront sử dụng để lưu trữ các bản sao đã được lưu trữ tại chỗ của nội dung của bạn gần khách hàng để giao hàng nhanh hơn**.

#### **Nguồn (Origin)**

Giả sử dữ liệu của công ty bạn được lưu trữ tại Brasil và bạn có khách hàng sống tại Trung Quốc. 

Để cung cấp nội dung cho những khách hàng này, bạn không cần phải di chuyển toàn bộ nội dung đến một trong các Khu vực Trung Quốc.

#### **Địa điểm Biên (Edge Location)**


Thay vì yêu cầu khách hàng lấy dữ liệu của họ từ Brasil, bạn có thể lưu trữ một bản sao tại một địa điểm biên gần khách hàng của bạn tại Trung Quốc.

#### **Khách hàng**

Khi một khách hàng tại Trung Quốc yêu cầu một trong các tệp của bạn, **Amazon CloudFront** sẽ lấy tệp từ bộ nhớ cache tại địa điểm biên và giao tệp đến khách hàng. 

Tệp được giao tới khách hàng nhanh hơn vì nó đến từ địa điểm biên gần Trung Quốc thay vì nguồn gốc ở Brasil.
