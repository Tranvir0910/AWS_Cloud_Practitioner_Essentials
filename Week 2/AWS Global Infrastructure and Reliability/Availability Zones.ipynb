### **Availability Zones**

- Một Khu vực Khả dụng là một trung tâm dữ liệu đơn lẻ hoặc một nhóm các trung tâm dữ liệu trong một Khu vực. 

- Các Khu vực Khả dụng được đặt cách xa nhau hàng chục dặm. Điều này đủ gần để có độ trễ thấp (thời gian giữa việc yêu cầu nội dung và nhận nội dung) giữa các Khu vực Khả dụng. 

- Tuy nhiên, nếu xảy ra một thảm họa tại một phần của Khu vực, chúng đủ xa để giảm khả năng nhiều Khu vực Khả dụng bị ảnh hưởng.

#### **Running Amazon EC2 instances in multiple Availability Zones**

##### **Amazon EC2 instance in a single Availability Zone**

Giả sử bạn đang chạy một ứng dụng trên một Amazon EC2 instances đơn lẻ trong **Khu vực Bắc California**. 

Instances đang chạy trong **Khu vực Khả dụng us-west-1a**. **Nếu us-west-1a gặp sự cố, bạn sẽ mất Instances** của mình.

##### **Amazon EC2 instances in multiple Availability Zones**

Một thực hành tốt là **chạy ứng dụng qua ít nhất hai Khu vực Khả dụng** trong một Khu vực. 

Trong ví dụ này, bạn có thể chọn chạy thêm một thể hiện Amazon EC2 thứ hai trong us-west-1b.

##### **Availability Zone failure**

Nếu **us-west-1a gặp sự cố**, ứng dụng của bạn vẫn sẽ **tiếp tục chạy trong us-west-1b**.
