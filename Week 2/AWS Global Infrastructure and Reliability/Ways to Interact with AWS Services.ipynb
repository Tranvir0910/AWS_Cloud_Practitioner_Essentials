### **Ways to Interact with AWS Services**

#### **AWS Management Console**

**Giao diện Quản lý AWS** là một giao diện dựa trên web cho việc truy cập và quản lý các dịch vụ AWS. Bạn có thể nhanh chóng truy cập các dịch vụ đã sử dụng gần đây và tìm kiếm các dịch vụ khác theo tên, từ khóa hoặc viết tắt. 

Bảng điều khiển bao gồm các trình hướng dẫn và quy trình làm việc tự động có thể đơn giản hóa quá trình hoàn thành các nhiệm vụ.

Bạn cũng có thể sử dụng ứng dụng di động AWS Console để thực hiện các nhiệm vụ như giám sát tài nguyên, xem cảnh báo và truy cập thông tin thanh toán. Nhiều định dạng danh tính có thể tiếp tục đăng nhập vào ứng dụng di động AWS Console cùng một lúc.

#### **Giao diện Dòng Lệnh AWS (AWS Command Line Interface) (CLI)**

Để tiết kiệm thời gian khi thực hiện các yêu cầu API, bạn có thể sử dụng **Giao diện Dòng Lệnh AWS (AWS CLI)**. AWS CLI cho phép bạn điều khiển nhiều dịch vụ AWS trực tiếp từ dòng lệnh trong một công cụ duy nhất. AWS CLI có sẵn cho người dùng trên Windows, macOS và Linux.

Bằng cách sử dụng AWS CLI, bạn có thể tự động hóa các hành động mà các dịch vụ và ứng dụng của bạn thực hiện thông qua các tập lệnh. 

Ví dụ, bạn có thể sử dụng các lệnh để khởi chạy một thể hiện Amazon EC2, kết nối một thể hiện Amazon EC2 với một nhóm Tăng tự động cụ thể và nhiều hơn nữa.

#### **Software development kits (SDKs)**

Một lựa chọn khác để truy cập và quản lý các dịch vụ AWS là các bộ phát triển phần mềm (SDKs). SDKs giúp bạn sử dụng dễ dàng các dịch vụ AWS thông qua một API được thiết kế cho ngôn ngữ lập trình hoặc nền tảng của bạn. SDKs cho phép bạn sử dụng các dịch vụ AWS với các ứng dụng hiện có của bạn hoặc tạo các ứng dụng hoàn toàn mới sẽ chạy trên AWS.

Để giúp bạn bắt đầu sử dụng SDKs, AWS cung cấp tài liệu và mã mẫu cho mỗi ngôn ngữ lập trình được hỗ trợ. Các ngôn ngữ lập trình được hỗ trợ bao gồm C++, Java, .NET và nhiều ngôn ngữ khác.
