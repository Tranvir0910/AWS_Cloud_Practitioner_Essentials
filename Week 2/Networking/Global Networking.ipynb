## **Global Networking**

### **Domain Name System (DNS)**

Bạn có thể tưởng tượng DNS như cuốn danh bạ điện thoại trên internet. Quá trình giải quyết DNS là quá trình dịch một tên miền thành một địa chỉ IP.

Ví dụ, giả sử bạn muốn truy visit trang web của Công Ty Bất Kỳ (AnyCompany).

- Khi bạn nhập tên miền vào trình duyệt, yêu cầu này được gửi đến một máy chủ DNS.

- Máy chủ DNS yêu cầu máy chủ web cung cấp địa chỉ IP tương ứng với trang web của Công Ty Bất Kỳ.

- Máy chủ web phản hồi bằng cách cung cấp địa chỉ IP cho trang web của Công Ty Bất Kỳ, 192.0.2.0.

### **Amazon Route 53**

**Amazon Route 53** là một dịch vụ web DNS. Nó cung cấp cho các nhà phát triển và doanh nghiệp một cách đáng tin cậy để định tuyến người dùng cuối đến các ứng dụng internet được lưu trữ trong AWS.

**Amazon Route 53** kết nối yêu cầu của người dùng đến cơ sở hạ tầng chạy trên AWS (**Amazon EC2 instances và load balancers**). Nó cũng có thể định tuyến người dùng đến cơ sở hạ tầng ngoài AWS.

Một tính năng khác của Route 53 là khả năng quản lý bản ghi DNS cho tên miền. 

Bạn có thể đăng ký tên miền mới trực tiếp trong Route 53. Bạn cũng có thể chuyển bản ghi DNS cho các tên miền hiện có được quản lý bởi các công ty đăng ký tên miền khác. 

Điều này cho phép bạn quản lý tất cả tên miền của mình tại một vị trí duy nhất.

**Ví dụ:** Cách Amazon Route 53 và Amazon CloudFront phân phối nội dung

Giả sử ứng dụng của Công Ty Bất Kỳ đang chạy trên một số phiên bản Amazon EC2. Những phiên bản này nằm trong một nhóm tự động co dãn (**Auto Scaling**) kết nối với một Bộ cân bằng tải Ứng dụng (**Application Load Balancer**).

- Một khách hàng yêu cầu dữ liệu từ ứng dụng bằng cách truy cập trang web của Công Ty Bất Kỳ.

- **Amazon Route 53** sử dụng giải quyết DNS để xác định địa chỉ IP tương ứng với AnyCompany.com, là 192.0.2.0. Thông tin này được gửi lại cho khách hàng.

- Yêu cầu của khách hàng được gửi đến vị trí cạnh gần nhất thông qua **Amazon CloudFront**.

- **Amazon CloudFront** kết nối với Bộ cân bằng tải Ứng dụng (**ALB**), sau đó gửi gói tin đầu vào đến một phiên bản **Amazon EC2**.
