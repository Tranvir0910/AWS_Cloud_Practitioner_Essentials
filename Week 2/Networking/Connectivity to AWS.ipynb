### **Amazon Virtual Private Cloud (Amazon VPC)**

Dịch vụ mạng có thể sử dụng để thiết lập ranh giới xung quanh các tài nguyên AWS của bạn là Amazon Virtual Private Cloud (Amazon VPC).

**Amazon VPC** cho phép bạn cung cấp một phần cách ly trong Đám Mây AWS. Trong phần cách ly này, bạn có thể khởi chạy các tài nguyên trong một mạng ảo mà bạn tự định nghĩa. 

Trong một Mạng riêng ảo (VPC), bạn có thể tổ chức các tài nguyên của mình vào các mạng con. Một mạng con là một phần của một VPC có thể chứa các tài nguyên như các máy chủ Amazon EC2."

#### **Internet gateway**

Để cho phép lưu lượng truy cập công cộng từ internet truy cập vào VPC của bạn, bạn đính kèm cổng internet vào VPC.

**Internet gateway** là một kết nối giữa một VPC và internet.

### **What if you have a VPC that includes only private resources?**

#### **Virtual private gateway**

Bạn có thể tưởng tượng internet như con đường giữa ngôi nhà của bạn và quán cà phê. Giả sử bạn đang đi trên con đường này cùng với một vệ sĩ để bảo vệ bạn. Bạn vẫn đang sử dụng cùng một con đường như các khách hàng khác, nhưng có thêm một lớp bảo vệ bổ sung.

Vệ sĩ tương tự như một kết nối mạng riêng ảo (VPN) mã hóa (hoặc bảo vệ) lưu lượng internet của bạn khỏi tất cả các yêu cầu khác xung quanh.

**Vitual Private Gateway** là thành phần cho phép lưu lượng internet được bảo vệ vào VPC. 

Mặc dù kết nối của bạn tới quán cà phê có thêm bảo vệ bổ sung, tình trạng ùn tắc giao thông vẫn có thể xảy ra vì bạn đang sử dụng cùng một con đường như các khách hàng khác.

**Virtual Private Gateway** cho phép bạn thiết lập một kết nối mạng riêng ảo (VPN) giữa VPC của bạn và một mạng riêng, chẳng hạn như một trung tâm dữ liệu tại chỗ hoặc mạng nội bộ của doanh nghiệp. 

Một cổng riêng ảo cho phép lưu lượng vào VPC chỉ khi nó đến từ một mạng được phê duyệt.

### **AWS Direct Connect**

**AWS Direct Connect** là một dịch vụ cho phép bạn thiết lập một kết nối riêng tư chuyên dụng giữa trung tâm dữ liệu của bạn và một VPC.

Giả sử có một tòa nhà căn hộ với một hành lang kết nối trực tiếp từ tòa nhà đến quán cà phê. Chỉ có cư dân của tòa nhà căn hộ mới có thể đi qua hành lang này.

Hành lang riêng tư này cung cấp cùng loại kết nối chuyên dụng như **AWS Direct Connect**. Cư dân có thể vào quán cà phê mà không cần sử dụng con đường công cộng chia sẻ với khách hàng khác.

Kết nối riêng mà AWS Direct Connect cung cấp giúp bạn giảm chi phí mạng và tăng lượng băng thông có thể truyền qua mạng của bạn.
