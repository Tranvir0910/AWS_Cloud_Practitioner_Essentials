## **Monolithic Applications and Microservices**

### **Monolithic Applications**

Ứng dụng được tạo thành từ nhiều thành phần. Các thành phần này giao tiếp với nhau để truyền dữ liệu, thực hiện các yêu cầu và duy trì hoạt động của ứng dụng.

Giả sử bạn có một ứng dụng với các thành phần được kết nối chặt chẽ. Những thành phần này có thể bao gồm cơ sở dữ liệu, máy chủ, giao diện người dùng, logic kinh doanh và những thứ khác. 

Loại kiến trúc này có thể được coi là một ứng dụng monolithic.

Nếu một thành phần duy nhất gặp sự cố, các thành phần khác cũng gặp sự cố và có thể toàn bộ ứng dụng bị hỏng.

### **Microservices**

Để giúp duy trì tính sẵn có của ứng dụng **khi một thành phần duy nhất gặp sự cố**, bạn có thể thiết kế ứng dụng của mình thông qua cách tiếp cận **microservices**.

Trong cách tiếp cận microservices, các thành phần của ứng dụng được **loosely coupled (kết nối lỏng)**.

Trong trường hợp này, nếu một thành phần gặp sự cố, các thành phần khác vẫn tiếp tục hoạt động vì họ đang giao tiếp với nhau. Việc kết nối lỏng lẻo ngăn chặn sự cố lan tỏa cho toàn bộ ứng dụng.

Khi thiết kế ứng dụng trên AWS, bạn có thể sử dụng cách tiếp cận microservices với các dịch vụ và thành phần thực hiện các chức năng khác nhau. 

Hai dịch vụ hỗ trợ tích hợp ứng dụng là **Amazon Simple Notification Service (Amazon SNS)** và **Amazon Simple Queue Service (Amazon SQS)**.
