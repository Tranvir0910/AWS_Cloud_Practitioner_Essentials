## **Amazon Simple Queue Service (Amazon SQS)**

**Amazon Simple Queue Service (Amazon SQS)** là một dịch vụ hàng đợi thông điệp.

Amazon SQS, bạn có thể gửi, lưu trữ và nhận thông điệp giữa các thành phần phần mềm mà không mất thông điệp hoặc yêu cầu các dịch vụ khác phải có sẵn.

Trong Amazon SQS, một ứng dụng gửi các thông điệp vào một hàng đợi. Người dùng hoặc dịch vụ lấy thông điệp từ hàng đợi, xử lý nó và sau đó xóa nó khỏi hàng đợi.

##### **Ví dụ**: **Xử lý đơn hàng**

Quán cà phê có quy trình đặt hàng với hai thành phần: nhân viên thu ngân và người pha chế. 

Nhưng khi họ không đồng bộ, quy trình trở nên chậm chạp và không hiệu quả. Chủ quán quyết định sử dụng hàng đợi để giải quyết vấn đề này. 

Nhân viên thu ngân đặt đơn hàng vào hàng đợi, người pha chế lấy đơn hàng từ hàng đợi để thực hiện. Nhờ cách tiếp cận này, quy trình đặt hàng trở nên hiệu quả hơn và khách hàng không cần chờ lâu để nhận đơn hàng.

##### **Ví dụ: Đơn hàng trong hàng đợi**

Một dịch vụ hàng đợi thông điệp như **Amazon SQS** cho phép các thông điệp giữa các thành phần của ứng dụng được kết nối lỏng lẻo **(decoupled)**.

- Khách hàng đặt đơn hàng với nhân viên thu ngân.

- Nhân viên thu ngân đưa đơn hàng vào hàng đợi, tạo thành một bảng đơn hàng hoạt động như bộ đệm giữa nhân viên thu ngân và người pha chế.

- Ngay cả khi người pha chế đang nghỉ hoặc bận với đơn hàng khác, nhân viên thu ngân vẫn có thể tiếp tục đặt các đơn hàng mới vào hàng đợi.

- Người pha chế kiểm tra hàng đợi và lấy đơn hàng từ hàng đợi để thực hiện.

- Người pha chế chuẩn bị đồ uống và giao cho khách hàng.

- Khi đã hoàn thành, người pha chế loại bỏ đơn hàng đã thực hiện khỏi hàng đợi.

- Trong khi người pha chế đang chuẩn bị đồ uống, nhân viên thu ngân có thể tiếp tục nhận đơn hàng mới và thêm chúng vào hàng đợi.
