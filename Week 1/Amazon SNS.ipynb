## **Amazon Simple Notification Service (Amazon SNS)**

Amazon Simple Notification Service (Amazon SNS) là một dịch vụ **publish/subscribe**.

Sử dụng các chủ đề **(topics)** trong **Amazon SNS**, người xuất bản **(publisher)** gửi thông điệp tới các người đăng ký **(subscribers)**.

Trong **Amazon SNS**, các người đăng ký có thể là các máy chủ web, địa chỉ email, các hàm AWS Lambda hoặc một số lựa chọn khác.

##### **Publishing updates from a single topic ( Xuất bản thông tin từ một chủ đề duy nhất )**

Giả sử quán cà phê có một bản tin duy nhất bao gồm cập nhật từ tất cả các lĩnh vực kinh doanh của mình. Nó bao gồm các chủ đề như phiếu giảm giá, thú vị về cà phê và sản phẩm mới. **Tất cả các chủ đề này được nhóm lại vì đây là một bản tin duy nhất**. Tất cả khách hàng đăng ký nhận bản tin sẽ nhận được cập nhật về phiếu giảm giá, thú vị về cà phê và sản phẩm mới.

Sau một thời gian, một số khách hàng cho biết họ muốn nhận các bản tin riêng biệt cho chỉ các chủ đề cụ thể mà họ quan tâm. Chủ cửa hàng quyết định thử tiếp cận này.

#### **Publishing updates from multiple topics ( Xuất bản thông tin từ nhiều chủ đề )**

Bây giờ, thay vì có một bản tin duy nhất cho tất cả các chủ đề, quán cà phê đã chia nó thành ba bản tin riêng biệt. 

Mỗi bản tin dành riêng cho một chủ đề cụ thể: phiếu giảm giá, thú vị về cà phê và sản phẩm mới.

Người đăng ký sẽ nhận ngay lập tức cập nhật chỉ cho các chủ đề cụ thể mà họ đã đăng ký.

Có thể người đăng ký đăng ký cho một chủ đề duy nhất hoặc nhiều chủ đề. 

Ví dụ, khách hàng đầu tiên đăng ký chỉ cho chủ đề phiếu giảm giá, và người đăng ký thứ hai đăng ký chỉ cho chủ đề thú vị về cà phê. Khách hàng thứ ba đăng ký cả hai chủ đề thú vị về cà phê và sản phẩm mới.
