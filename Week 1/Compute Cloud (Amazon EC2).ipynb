### **Amazon Elastic Compute Cloud (Amazon EC2)**

[Amazon Elastic Compute Cloud (Amazon EC2)](https://aws.amazon.com/ec2/)

Là dịch vụ của AWS cung cấp khả năng tính toán linh hoạt, có thể thay đổi kích thước và bảo mật trong đám mây thông qua các 
phiên bản máy chủ ảo **(Amazon EC2 instances)**.

Các ưu điểm của việc sử dụng **Amazon EC2** instance bao gồm:

- Triển khai và khởi chạy Amazon EC2 instance trong vài phút.

- Dừng sử dụng instance khi bạn đã hoàn tất công việc chạy một tải công việc.

- Bạn chỉ trả tiền cho thời gian tính toán bạn sử dụng khi một instance đang chạy, không phải khi nó dừng hoặc bị chấm dứt.

- Tiết kiệm chi phí bằng cách chỉ trả tiền cho khả năng máy chủ bạn cần hoặc muốn sử dụng.

**How Amazon EC2 works**

- **Khởi tạo (Launch)**: Chọn mẫu cấu hình cơ bản và loại instance. Đặt các thiết lập bảo mật.

- **Kết nối (Connect)**: Kết nối đến instance qua nhiều cách khác nhau, trao đổi dữ liệu.

- **Sử dụng (Use)**: Thực hiện các lệnh cài đặt, quản lý tệp, thêm lưu trữ, và nhiều nhiệm vụ khác.
