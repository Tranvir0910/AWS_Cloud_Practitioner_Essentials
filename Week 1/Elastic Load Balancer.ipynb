## **Elastic Load Balancer (ELB)**

**Elastic Load Balancing** là dịch vụ của AWS giúp **tự động phân phối lưu lượng ứng dụng** đến nhiều nguồn tài nguyên khác nhau, chẳng hạn như các Amazon EC2 instances

**Bộ cân bằng tải (load balancer)** hoạt động như một điểm liên lạc duy nhất cho toàn bộ lưu lượng web đến nhóm Auto Scaling của bạn.

Điều này có nghĩa là khi bạn **thêm hoặc loại bỏ các Amazon EC2 íntances** tùy theo lưu lượng truy cập đến, các yêu cầu sẽ được định tuyến đến bộ cân bằng tải trước tiên. 

Sau đó, các yêu cầu sẽ **được phân phối đến nhiều nguồn tài nguyên sẽ xử lý chúng**. 


**Ví dụ:** 

Nếu bạn có nhiều thể hiện Amazon EC2, Elastic Load Balancing sẽ phân phối công việc đều đặn giữa các thể hiện này để đảm bảo không có một thể hiện nào phải chịu toàn bộ tải.

**Elastic Load Balancing** và **Amazon EC2 Auto Scaling** là hai dịch vụ riêng biệt, nhưng chúng hoạt động cùng nhau để đảm bảo các ứng dụng chạy trên Amazon EC2 có thể cung cấp hiệu suất và khả năng sẵn sàng cao.
