### **Cloud Computing**





Điện toán đám mây là việc cung cấp tài nguyên CNTT theo 
yêu cầu qua Internet với mức giá trả theo mức sử dụng.

Cung cấp theo yêu cầu cho thấy AWS có 
các tài nguyên bạn cần khi bạn cần.

### **What is a Client-Server Model**


Bạn vừa tìm hiểu thêm về AWS và cách hầu hết tất cả điện toán hiện đại sử dụng mô hình máy khách-máy chủ cơ bản. Hãy tóm tắt lại mô hình máy khách-máy chủ là gì.

Trong điện toán, **Client** có thể là một trình duyệt web hoặc ứng dụng dành cho máy tính để bàn mà một người tương tác để gửi yêu cầu đến máy chủ. 

**Sever** có thể là các dịch vụ như **Amazon Elastic Compute Cloud (Amazon EC2)**, một loại máy chủ ảo.

**Ví dụ:**

Giả sử khách hàng đưa ra yêu cầu về một bài báo, tỷ số trong một trò chơi trực tuyến hoặc một video hài hước. Máy chủ đánh giá các chi tiết của yêu cầu này và đáp ứng nó bằng cách trả lại thông tin cho máy khách.

### **Deployment Models for Cloud Computing ( Các mô hình triển khai cho điện toán đám mây )**

Ba mô hình triển khai điện toán đám mây là: **cloud-based, on-premises, and hybrid**.

#### **Cloud-based Deployment ( Triển khai trên đám mây )** 

- Run all parts of the application in the cloud. ( Chạy tất cả các phần của ứng dụng trên đám mây )

- Migrate existing applications to the cloud. ( Di chuyển các ứng dụng hiện có sang đám mây )

- Design and build new applications in the cloud. ( Thiết kế và xây dựng các ứng dụng mới trên đám mây )

Trong **Cloud-based Deployment Model**, bạn có thể di chuyển các ứng dụng hiện có sang đám mây hoặc bạn có thể thiết kế và xây dựng các ứng dụng mới trên đám mây. 

Bạn có thể xây dựng các ứng dụng đó trên cơ sở hạ tầng cấp thấp yêu cầu nhân viên CNTT của bạn quản lý chúng. 

Ngoài ra, bạn có thể xây dựng chúng bằng cách sử dụng các dịch vụ cấp cao hơn giúp giảm các yêu cầu về quản lý, kiến ​​trúc và mở rộng quy mô của cơ sở hạ tầng cốt lõi. 

Ví dụ: một công ty có thể tạo một ứng dụng bao gồm các máy chủ ảo, cơ sở dữ liệu và các thành phần mạng hoàn toàn dựa trên đám mây.

#### **On-premises Deployment ( Triển khai tại chổ )**

- Deploy resources by using virtualization and resource management tools. (Triển khai tài nguyên bằng cách sử dụng các công cụ ảo hóa và quản lý tài nguyên )

- Increase resource utilization by using application management and virtualization technologies. ( Tăng mức sử dụng tài nguyên bằng cách sử dụng các công nghệ ảo hóa và quản lý ứng dụng )

**On-premises deployment** còn được gọi là private cloud deployment ( Triển khai đám mây riêng tư ). 

Trong mô hình này, các tài nguyên được triển khai tại cơ sở bằng cách sử dụng các công cụ quản lý tài nguyên và ảo hóa.

Có rất ít hoặc không có sự khác biệt giữa mô hình public và mô hình private từ quan điểm kỹ thuật, vì kiến ​​trúc của chúng rất giống nhau. 

Tuy nhiên, trái ngược với public clouds có sẵn cho người dùng, private clouds chỉ có một công ty cụ thể sở hữu đám mây riêng. 

Đó là lý do tại sao nó còn được gọi là **mô hình nội bộ (internal)** hoặc **mô hình công ty (corporate)**.

#### **Hybrid Deployment ( Triển khai kết hợp )**

- Connect cloud-based resources to on-premises infrastructure. ( Kết nối tài nguyên dựa trên đám mây với cơ sở hạ tầng tại chỗ. )

- Integrate cloud-based resources with legacy IT applications. ( Tích hợp các tài nguyên dựa trên đám mây với các ứng dụng CNTT cũ. )

Trong **Hybrid Deployment**, tài nguyên dựa trên đám mây được **kết nối on-premises infrastructure**. 

Bạn có thể muốn sử dụng phương pháp này trong một số tình huống. 

**Ví dụ:** 

Bạn có các ứng dụng cũ được duy trì tốt hơn tại cơ sở hoặc các quy định của chính phủ yêu cầu doanh nghiệp của bạn lưu giữ một số hồ sơ nhất định tại cơ sở.

**Ví dụ:** 

Giả sử rằng một công ty muốn sử dụng các dịch vụ đám mây có thể tự động hóa quá trình xử lý và phân tích dữ liệu hàng loạt. 

Tuy nhiên, công ty có một số ứng dụng cũ phù hợp hơn tại cơ sở và sẽ không được chuyển sang đám mây. 

Với triển khai kết hợp, công ty sẽ có thể giữ các ứng dụng cũ tại cơ sở đồng thời hưởng lợi từ các dịch vụ dữ liệu và phân tích chạy trên đám mây.

### **Benefits of Cloud Computing**

#### **Trade upfront expense for variable expense ( Trao đổi chi phí ban đầu thành chi phí biến đổi )**

- Cloud Computing cho phép doanh nghiệp áp dụng mô hình chi trả chi phí biến đổi, nghĩa là **chỉ trả tiền cho các tài nguyên máy tính mà họ sử dụng**, thay vì đầu tư mạnh vào các trung tâm dữ liệu và máy chủ trước khi biết cách sử dụng chúng. 

- Điều này giúp doanh nghiệp triển khai các giải pháp sáng tạo trong khi tiết kiệm chi phí.

#### **Stop spending money to run and maintain data centers ( Ngừng chi tiêu để vận hành và bảo trì trung tâm dữ liệu )**

- Việc tính toán tại các trung tâm dữ liệu thường đòi hỏi bạn phải tiêu nhiều tiền và thời gian để quản lý cơ sở hạ tầng và máy chủ. 

- Lợi ích của Cloud Computing là bạn có thể tập trung ít hơn vào các nhiệm vụ này và tập trung hơn vào ứng dụng và khách hàng của bạn.

#### **Stop guessing capacity ( Ngừng đoán dung lượng )**

- Với Cloud Computing, **bạn không phải dự đoán bao nhiêu dung lượng cơ sở hạ tầng** bạn sẽ cần trước khi triển khai một ứng dụng. 

- Thay vào đó, bạn có thể khởi chạy các phiên bản Amazon EC2 khi cần thiết và chỉ trả tiền cho thời gian tính toán bạn sử dụng. 

- Thay vì trả tiền cho các tài nguyên không sử dụng hoặc phải đối mặt với dung lượng hạn chế, bạn chỉ sử dụng dung lượng cần thiết. 

- Bạn cũng có thể mở rộng hoặc thu hẹp quy mô phù hợp với nhu cầu.

#### **Benefit from massive economies of scale ( Hưởng lợi từ quy mô kinh tế khổng lồ )**

- Sử dụng Cloud Computing, **bạn có thể đạt được chi phí biến đổi thấp hơn so với việc tự mình xây dựng cơ sở hạ tầng**. 

- Bởi vì việc sử dụng của hàng trăm nghìn khách hàng có thể tập hợp trong đám mây, nhà cung cấp như AWS có thể đạt được quy mô kinh tế lớn hơn. 

- Quy mô kinh tế dẫn đến giá thấp hơn khi thanh toán theo tiêu chí "trả tiền khi sử dụng".

#### **Increase speed and agility ( Tăng tốc độ và linh hoạt )**

- Sự linh hoạt của Cloud Computing làm cho việc phát triển và triển khai ứng dụng dễ dàng hơn. 

- Tính linh hoạt này cung cấp cho bạn nhiều thời gian để thử nghiệm và đổi mới. 

- Khi tính toán tại các trung tâm dữ liệu, có thể mất hàng tuần để có được các tài nguyên mới bạn cần. So sánh với điều này, Cloud Computing cho phép bạn truy cập tài nguyên mới trong vòng vài phút.

#### **Go global in minutes ( Phủ sóng toàn cầu trong vài phút )**

- Phạm vi toàn cầu của Đám mây AWS cho phép bạn triển khai ứng dụng đến khách hàng trên khắp thế giới một cách nhanh chóng, đồng thời cung cấp cho họ độ trễ thấp. 

- Điều này có nghĩa là ngay cả khi bạn đặt ở một nơi khác so với khách hàng của mình, họ vẫn có thể truy cập ứng dụng của bạn với độ trễ tối thiểu.
