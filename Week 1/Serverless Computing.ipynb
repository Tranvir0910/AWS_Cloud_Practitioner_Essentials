## **Serverless Computing**

#### **Serverless Computing**

Amazon EC2 là một dịch vụ cho phép bạn chạy các máy chủ ảo trong đám mây. Nếu bạn có các ứng dụng muốn chạy trên Amazon EC2, bạn phải thực hiện các bước sau:

- Cấp phát các instances (máy chủ ảo).

- Tải lên mã của bạn.

- Tiếp tục quản lý các instances trong khi ứng dụng của bạn đang chạy.

**"Serverless"** có nghĩa là mã của bạn chạy trên các máy chủ, nhưng **bạn không cần phải cấu hình hoặc quản lý những máy chủ này**. 

Với tính toán không cần máy chủ, bạn có thể tập trung nhiều hơn vào việc đổi mới các sản phẩm và tính năng mới thay vì duy trì máy chủ.

Lợi ích khác của tính toán không cần máy chủ là tính linh hoạt trong việc tự động mở rộng các ứng dụng không cần máy chủ. 

Tính toán không cần máy chủ có thể điều chỉnh khả năng của ứng dụng bằng cách sửa đổi các đơn vị tiêu thụ, chẳng hạn như thông lượng và bộ nhớ.

Một dịch vụ của AWS cho tính toán không cần máy chủ là **AWS Lambda**.

#### **AWS Lambda**

**AWS Lambda** là một dịch vụ cho phép bạn chạy mã mà không cần phải cấu hình hoặc quản lý máy chủ.

Khi sử dụng AWS Lambda, bạn chỉ trả tiền cho thời gian tính toán bạn tiêu thụ. 

Phí áp dụng chỉ khi mã của bạn đang chạy. Bạn cũng có thể chạy mã cho bất kỳ loại ứng dụng hoặc dịch vụ backend nào mà không cần quản lý.

**Ví dụ**: 

Một Lambda function đơn giản có thể là việc tự động thay đổi kích thước hình ảnh được tải lên lên đám mây AWS. Trong trường hợp này, hàm sẽ kích hoạt khi có một hình ảnh mới được tải lên.

#### **How AWS Lambda works**

- Bạn tải mã của mình lên Lambda.

- Bạn thiết lập mã của mình để kích hoạt từ một nguồn sự kiện, chẳng hạn như các dịch vụ AWS, ứng dụng di động hoặc các điểm cuối HTTP.

- Lambda chỉ chạy mã của bạn khi có sự kích hoạt từ nguồn sự kiện.

- Bạn chỉ trả tiền cho thời gian tính toán bạn sử dụng. Trong ví dụ trước về việc thay đổi kích thước hình ảnh, bạn chỉ trả tiền cho thời gian tính toán khi tải lên hình ảnh mới. Việc tải lên hình ảnh kích hoạt Lambda chạy mã cho chức năng thay đổi kích thước hình ảnh.

#### **Containers**

Trong AWS, bạn cũng có thể xây dựng và chạy các ứng dụng được đóng gói bằng container.

**Container** cung cấp cho bạn một cách tiêu chuẩn để đóng gói mã ứng dụng và các thành phần phụ thuộc vào một đối tượng duy nhất. 

Bạn cũng có thể sử dụng **container** cho các quy trình và luồng công việc có yêu cầu cơ bản về bảo mật, đáng tin cậy và khả năng mở rộng.

##### **Ví dụ: Một máy chủ với nhiều container**

Giả sử một nhà phát triển ứng dụng của công ty có môi trường trên máy tính của họ khác với môi trường trên các máy tính được sử dụng bởi nhân viên IT. 

Nhà phát triển muốn đảm bảo rằng môi trường của ứng dụng được duy trì nhất quán bất kể khi triển khai, vì vậy họ sử dụng phương pháp container. 

Điều này giúp giảm thời gian dành cho gỡ lỗi ứng dụng và chẩn đoán sự khác biệt trong môi trường tính toán.

##### **Ví dụ: Hàng chục máy chủ với hàng trăm container**

Khi chạy các ứng dụng được đóng gói bằng container, việc xem xét khả năng mở rộng là rất quan trọng. 

Giả sử thay vì một máy chủ duy nhất với nhiều container, bạn phải quản lý hàng chục máy chủ với hàng trăm container. 

Hoặc có thể bạn phải quản lý hàng trăm máy chủ với hàng nghìn container. 

Ở quy mô lớn như vậy, hãy tưởng tượng xem việc giám sát việc sử dụng bộ nhớ, bảo mật, ghi nhật ký và các yêu cầu khác sẽ mất bao nhiêu thời gian.

#### **Amazon ECS**

Dịch vụ Quản lý Container Đàn Hồi của Amazon **(Amazon Elastic Container Service - Amazon ECS)** là một hệ thống quản lý container có khả năng mở rộng và hiệu suất cao, cho phép bạn chạy và mở rộng các ứng dụng được đóng gói bằng container trên AWS.

**Amazon ECS** hỗ trợ các container của Docker. Docker là một nền tảng phần mềm cho phép bạn xây dựng, kiểm tra và triển khai các ứng dụng nhanh chóng. 

AWS hỗ trợ việc sử dụng Docker Community Edition mã nguồn mở và Docker Enterprise Edition có tính phí. 

Với Amazon ECS, bạn có thể sử dụng các API calls để khởi chạy và dừng các ứng dụng hỗ trợ Docker.
