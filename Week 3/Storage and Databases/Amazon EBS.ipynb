### **Instance Stores and Amazon Elastic Block Store (Amazon EBS)**

#### **Instance stores**

Ổ lưu trữ của máy ảo cung cấp lưu trữ tạm thời ở mức khối cho một máy ảo Amazon EC2. 

Ổ lưu trữ của máy ảo là lưu trữ đĩa được gắn vật lý vào máy tính chủ cho một máy ảo EC2, do đó có tuổi thọ giống như máy ảo. 

**Khi máy ảo bị chấm dứt, bạn mất mọi dữ liệu trên ổ lưu trữ của máy ảo.**

**Một máy ảo Amazon EC2 có ổ lưu trữ của máy ảo được gắn kèm đang chạy.**

**Máy ảo bị tắt hoặc chấm dứt.**

**Tất cả dữ liệu trên ổ lưu trữ của máy ảo được gắn kèm sẽ bị xóa.**

#### **Ổ Lưu trữ Elastic Block của Amazon (Amazon EBS)**

**Amazon Elastic Block Store (Amazon EBS)** là một dịch vụ cung cấp các ổ lưu trữ cấp khối mà bạn có thể sử dụng với các máy ảo Amazon EC2. Nế**u bạn dừng hoặc chấm dứt một máy ảo Amazon EC2, tất cả dữ liệu trên ổ lưu trữ EBS đính kèm vẫn có sẵn.**

Để tạo một ổ EBS, bạn xác định cấu hình (như kích thước và loại ổ) và cung cấp nó. Sau khi bạn tạo một ổ EBS, nó có thể được đính kèm vào một máy ảo Amazon EC2.

Vì ổ lưu trữ EBS dành cho dữ liệu cần phải tồn tại, việc sao lưu dữ liệu là quan trọng. Bạn có thể thực hiện sao lưu tăng dần của các ổ EBS bằng cách tạo bản chụp Amazon EBS.

#### **Amazon EBS Snapshots**

**Amazon EBS Snapshots** là **một bản sao lưu tăng dần**. Điều này có nghĩa là **bản sao lưu đầu tiên của một ổ sao lưu toàn bộ dữ liệu**. 

Đối với các sao lưu sau này, chỉ có các khối dữ liệu đã thay đổi kể từ bản chụp gần đây nhất được lưu lại.

Sao lưu tăng dần khác với sao lưu toàn bộ, trong đó toàn bộ dữ liệu trong ổ lưu trữ được sao lưu mỗi lần thực hiện sao lưu. Sao lưu toàn bộ bao gồm dữ liệu không thay đổi kể từ lần sao lưu gần đây nhất.
