### **Amazon Simple Storage Service (Amazon S3)**

#### **Object Storage**

Trong lưu trữ đối tượng, mỗi đối tượng bao gồm dữ liệu, siêu dữ liệu và một khóa.

Dữ liệu có thể là hình ảnh, video, tài liệu văn bản hoặc bất kỳ loại tệp nào khác. Siêu dữ liệu chứa thông tin về dữ liệu là gì, cách sử dụng, kích thước đối tượng và vân vân. Khóa của đối tượng là định danh duy nhất của nó.

Hãy nhớ rằng khi bạn sửa đổi một tệp trong lưu trữ khối, chỉ có các phần thay đổi được cập nhật. Khi một tệp trong lưu trữ đối tượng được sửa đổi, toàn bộ đối tượng được cập nhật.

#### **Amazon Simple Storage Service (Amazon S3)**

**Amazon S3** là một dịch vụ cung cấp lưu trữ cấp đối tượng. Amazon S3 lưu trữ dữ liệu dưới dạng các đối tượng trong các thùng.

Ví dụ, bạn có thể sử dụng **Amazon S3** để lưu trữ các tệp sao lưu, tệp phương tiện cho một trang web hoặc tài liệu lưu trữ. 

**Amazon S3** cung cấp không gian lưu trữ không giới hạn. Kích thước tệp tối đa cho một đối tượng trong **Amazon S3 là 5 TB.**

Khi bạn tải lên một tệp lên Amazon S3, bạn có thể đặt quyền để kiểm soát khả năng nhìn thấy và truy cập vào nó. 

Bạn cũng có thể sử dụng tính năng phiên bản Amazon S3 để theo dõi các thay đổi đối với đối tượng của bạn theo thời gian.

#### **Amazon S3 Storage Classes**

Với Amazon S3, bạn chỉ trả tiền cho những gì bạn sử dụng. Bạn có thể chọn từ một loạt các lớp lưu trữ để chọn lựa phù hợp với nhu cầu kinh doanh và chi phí của bạn. Khi chọn một lớp lưu trữ Amazon S3, hãy xem xét hai yếu tố sau:

- Tần suất bạn dự định truy xuất dữ liệu của bạn

- Bạn cần dữ liệu của bạn có sẵn như thế nào

- **Amazon S3 Standard:** Dữ liệu truy cập thường xuyên.

- **Amazon S3 Standard-IA:** Dữ liệu ít truy cập, giá lưu trữ thấp.

- **Amazon S3 One Zone-IA:** Dữ liệu ít truy cập, lưu trữ trong một Khu vực.

- **Amazon S3 Intelligent-Tiering:** Theo dõi mẫu truy cập và di chuyển tự động.

- **Amazon S3 Glacier Instant Retrieval:** Dữ liệu lưu trữ cần truy cập ngay.

- **Amazon S3 Glacier Flexible Retrieval:** Lưu trữ chi phí thấp cho dữ liệu lưu trữ.

- **Amazon S3 Glacier Deep Archive:** Lưu trữ lâu dài, chi phí thấp nhất.

- **Amazon S3 Outposts:** Lưu trữ đối tượng cho môi trường AWS Outposts.

### **Comparing Amazon EBS and Amazon S3**

**Amazon S3 (Simple Storage Service):**

- Loại lưu trữ đối tượng dựa trên đám mây.

- Dành cho việc lưu trữ và quản lý các đối tượng như tệp tin, hình ảnh, video, tài liệu, v.v.

- Cung cấp khả năng lưu trữ không giới hạn và đối tượng cá nhân có thể lên đến 5.000 gigabyte trong kích thước.

- Được sử dụng cho các trường hợp sử dụng đọc nhiều lần hoặc ghi một lần, như lưu trữ dữ liệu phân tán, tài liệu chia sẻ và hỗ trợ sao lưu.

- Đã được cung cấp kết nối web sẵn có, mỗi đối tượng có URL có thể kiểm soát quyền truy cập.

- Độ bền cao, với khả năng đạt tới 99.999999999% độ bền.

- Tích hợp khái niệm vùng lưu trữ để đảm bảo sự phân phối và khả năng dự phòng.

**Amazon EBS (Elastic Block Storage):**

- Loại lưu trữ khối dành cho máy ảo trên EC2 trong môi trường đám mây.

- Dành cho việc kết nối và sử dụng như ổ cứng thường trên máy tính.

- Cung cấp khả năng lưu trữ khối với kích thước lên đến 16 tebibytes và hỗ trợ lưu trữ thể rắn (SSD) và đĩa cứng quay.

- Thích hợp cho các trường hợp sử dụng cần sự thay đổi thường xuyên và đọc/ghi dữ liệu chi tiết tại mức khối, chẳng hạn như cơ sở dữ liệu, ứng dụng có yêu cầu hiệu suất cao.

- Không có sẵn URL, không được thiết kế cho truy cập trực tiếp qua web.

- Độ bền thấp hơn so với S3, tùy thuộc vào loại ổ cứng được sử dụng.

- Cần liên kết với máy ảo EC2 để sử dụng, tức là cần tài nguyên máy ảo để hoạt động.

**Sử dụng Amazon S3 khi:**

- **Lưu trữ đối tượng:**         Khi bạn cần lưu trữ và quản lý các tệp tin, hình ảnh, video, dữ liệu và đối tượng khác.

- **Tính sẵn sàng và phân phối:** Khi bạn cần có khả năng truy cập đến các đối tượng từ bất kỳ nơi nào thông qua kết nối web sẵn có.

- **Dung lượng không giới hạn:**  Khi bạn cần một không gian lưu trữ mở rộng linh hoạt để lưu trữ một lượng lớn dữ liệu.

- **Sao lưu và phục hồi:**        Khi bạn cần một giải pháp sao lưu dữ liệu dự phòng và khả năng phục hồi trong trường hợp sự cố.



**Sử dụng Amazon EBS khi:**

- **Hiệu suất và chỉnh sửa thường xuyên:** Khi bạn cần hiệu suất cao cho các ứng dụng yêu cầu đọc/ghi dữ liệu thường xuyên, chẳng hạn như cơ sở dữ liệu hoặc máy chủ ứng dụng.

- **Lưu trữ dữ liệu chi tiết tại mức khối:** Khi bạn cần làm việc với dữ liệu tại mức khối và thường xuyên thay đổi dữ liệu tại mức này, chẳng hạn như chỉnh sửa các tệp video hoặc tệp hình ảnh lớn.

- **Đáng tin cậy cho máy ảo EC2:** Khi bạn cần một ổ đĩa gắn kèm với máy ảo EC2 để lưu trữ dữ liệu riêng biệt cho máy ảo đó.

- **Hỗ trợ cho các tùy chỉnh hệ điều hành và ứng dụng:** Khi bạn cần tạo ra và quản lý các ổ đĩa tùy chỉnh dành riêng cho các mục đích nhất định trên máy ảo EC2.
